```python
import requests
import json

result = json.loads(requests.get('https://api.lyrics.ovh/v1/They Might Be Giants/Birdhouse in your soul').text)

#File name for path
filename = 'birdhouse_lyric.json'
with open(filename, 'w') as file: 
    json.dump(result, file, indent=4)

print(f"Data succesfully written to {filename}")    
```

## Question 2
2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob


nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

# Read in JSON file
with open('birdhouse_lyric.json', 'r') as file:
    data = json.load(file)

# Get lyrics
lyrics = data.get('lyrics', '')
print(lyrics)     

doc = nlp(lyrics)
polarity = doc._.blob.polarity

print(f"Song polarity score:", polarity)

#Sentiment of lyrics is a more positive connotation, as 0.045 is close to 1 than -1. 
```

## Question 3
3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
# Define function to fetch data and save to json file
def get_song(artist, song, filename):
    # Construct the API URL
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
    
    # Make the API request to fetch song lyrics
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        lyrics = data.get('lyrics', 'Lyrics not found')

        # Write the song details and lyrics to the file
        with open(filename, 'w') as file:
            json.dump({'artist': artist, 'song': song, 'lyrics': lyrics}, file, indent=4)

        print(f"Lyrics for '{song}' by {artist} have been written to {filename}")
    else:
        print(f"Failed to retrieve lyrics for '{song}' by {artist}. Status code: {response.status_code}")

# Test the function with four songs
get_song('The Beatles', 'Blackbird', 'beatles_blackbird_lyrics.json')
get_song('Taylor Swift', 'The Man', 'taylor_the_man_lyrics.json')
get_song('Chappell Roan', 'Hot to go!', 'chappell_hotogo_lyrics.json')
get_song('Dashboard Confessional', 'Stolen', 'dashboard_stolen_lyrics.json')
```
